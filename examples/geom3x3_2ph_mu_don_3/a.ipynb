```python

import numpy as np
import pickle
import scipy.io
import matplotlib.pyplot as plt
```

```python
filepath = '1x2.pkl'
with open(filepath, 'rb') as filepath:
    arquivos = pickle.load(filepath)

contour_points = arquivos['contour_points']
X = arquivos['outside_points']
```

```python
print(contour_points.shape)
```

```python
indx = np.where(contour_points[:,1]>=150)[0]
contour_points2 = contour_points[indx]
indx = np.where(contour_points2[:,1]<=250)[0]
contour_points2 = contour_points2[indx]

addx = np.arange(151,250)[:,np.newaxis]
addy = 150*np.ones(addx.shape)
contour_points2 = np.concatenate((contour_points2, np.concatenate((addx, addy), axis=1)), axis=0)
addy = 250*np.ones(addx.shape)
contour_points2 = np.concatenate((contour_points2, np.concatenate((addx, addy), axis=1)), axis=0)

addx = np.arange(401,500)[:,np.newaxis]
addy = 150*np.ones(addx.shape)
contour_points2 = np.concatenate((contour_points2, np.concatenate((addx, addy), axis=1)), axis=0)
addy = 250*np.ones(addx.shape)
contour_points2 = np.concatenate((contour_points2, np.concatenate((addx, addy), axis=1)), axis=0)
contour_points2[:,1] = contour_points2[:,1]-150
```

```python
plt.scatter(x=contour_points2[:,0], y=contour_points2[:,1])
```
