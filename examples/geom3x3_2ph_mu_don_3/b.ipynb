```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mus = .1
muq = .0025
mur = mus/muq
```

```python
def __nonlinear_scaler(mu):
    mus = .1
    muq = .0025
    mur = mus/muq
    
    yu =  __nonlinear_scaler_equation(mu, mus, muq, mur, multiplier=1)    
    yv =  __nonlinear_scaler_equation(mu, mus, muq, mur, multiplier=1)                  
    return 1*yu, 1*yv

def __nonlinear_scaler_equation(mu, mus, muq, mur, multiplier):
    mur = mur*multiplier
    
    a = mur / ( (mus**(-2) - muq**(-2)) +1 )
    b = 1- a * muq**(-2)
    
    return a*mu**(-2) +b
```

```python
x = np.arange(muq, mus+.0005, .001)
y, _ = __nonlinear_scaler(x)

print(__nonlinear_scaler(.1)[0])
```
