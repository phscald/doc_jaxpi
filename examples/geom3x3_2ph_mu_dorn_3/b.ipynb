```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mus = .1
muq = .0025
mur = mus/muq
```

```python
def __linear_scaler_equation(mu, mus, muq, mur):
    return (mur-1)/(mus-muq) * (mu-muq) +1

def __linear_scaler(mu):
        mus = .1
        muq = .0025
        mur = mus/muq
        
        yu =  __linear_scaler_equation(mu, mus, muq, mur)    
        yv =  __linear_scaler_equation(mu, mus, muq, mur) # 54/40              
        return yu, yv

def __nonlinear_scaler(mu):
    mus = .1
    muq = .0025
    mur = mus/muq
    
    yu =  __nonlinear_scaler_equation(mu, mus, muq, mur, multiplier=1)   
    yv =  __nonlinear_scaler_equation(mu, mus, muq, mur, multiplier=1)                  
    return 1*yu, 1*yv

def __nonlinear_scaler_equation(mu, mus, muq, mur, multiplier):
    mur = mur*multiplier
    
    a = mur / ( (mus**(-2) - muq**(-2)) +1 )
    b = 1- a * muq**(-2)
    
    return a*mu**(-2) +b

def __scale_function(funct, old_scale, new_scale):
    funct = funct -1 
    funct = funct/(old_scale-1)
    funct = funct*(new_scale-1)
    return funct+1
```

```python
x = np.arange(muq, mus+.0005, .001)
y, _ = __linear_scaler(x)
y = __scale_function(y, 40, 16)

print(__linear_scaler(.0025)[0])
```
