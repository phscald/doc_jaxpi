```python
import pickle
import numpy as np 
from matplotlib import pyplot as plt
```

```python
filepath = 'steady_3x3.pkl'
with open(filepath, 'rb') as filepath:
	arquivos = pickle.load(filepath)
print(arquivos.keys())
```

```python
print(np.arange(0,12.1))
```

```python
contour_points = arquivos['contour_points']
outside_points = arquivos['outside_points']
inside_points = arquivos['inside_points']
```

```python
print(contour_points.shape)
print(outside_points.shape)
print(inside_points.shape)
```

```python
print(np.max(outside_points[:,0])) # domain - black
print(np.max(inside_points[:,0])) # not domain - white
print(np.min(outside_points[:,0]))
print(np.min(inside_points[:,0]))
```

```python
print(np.max(outside_points[:,1])) # domain - black
print(np.max(inside_points[:,1])) # not domain - white
print(np.min(outside_points[:,1]))
print(np.min(inside_points[:,1]))
```

```python
plt.scatter(x=outside_points[:,0], y=outside_points[:,1], s=5)
```
